# Bar Chart

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_used_cars=pd.read_csv("bar_chart_data.csv")
df_used_cars
```

```python
plt.bar(x=df_used_cars["Brand"],
       height=df_used_cars["Cars Listings"])
plt.show()
```

```python
plt.figure(figsize=(9,6))
plt.bar(x=df_used_cars["Brand"],
       height=df_used_cars["Cars Listings"])
plt.show()
```

```python
plt.figure(figsize=(9,6))
plt.bar(x=df_used_cars["Brand"],
       height=df_used_cars["Cars Listings"])
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(9,6))
plt.bar(x=df_used_cars["Brand"],
       height=df_used_cars["Cars Listings"],
        color=["r","g","b","w","y","m","c"])
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(9,6))
plt.bar(x=df_used_cars["Brand"],
       height=df_used_cars["Cars Listings"],
        color="midnightblue")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(9,6))
plt.bar(x=df_used_cars["Brand"],
       height=df_used_cars["Cars Listings"],
        color="midnightblue")
plt.xticks(rotation=45, fontsize=13)
plt.yticks(fontsize=13)
plt.title("Cars listings by Brands", fontsize=16, fontweight="bold")
plt.ylabel("Number of Listings",fontsize=13)
plt.show()
```

# Pie Chart

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_fuel_engine_type = pd.read_csv("pie_chart_data.csv")
df_fuel_engine_type
```

```python
plt.figure(figsize=(10,8))
plt.pie(df_fuel_engine_type['Number of Cars'],
       labels=df_fuel_engine_type['Engine Fuel Type'].values,
       autopct='%.2f%%')
plt.show()
```

```python
sns.set_palette('colorblind')
plt.figure(figsize=(10,8))
plt.pie(df_fuel_engine_type['Number of Cars'],
       labels=df_fuel_engine_type['Engine Fuel Type'].values,
       autopct='%.2f%%',
       textprops={'size':'x-large',
                 'fontweight':'bold',
                 'rotation':'30',
                 'color':'w'})
plt.title('Cars by Engine Fuel Type',
         fontsize=18,
         fontweight='bold')
plt.legend()
plt.show()
```

# Stacked Area Chart

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_fuel_engine_types = pd.read_csv("stacked_area_chart_data.csv")
df_fuel_engine_types
```

```python
plt.stackplot(df_fuel_engine_types["Year"],
             df_fuel_engine_types["Diesel"],
             df_fuel_engine_types["Petrol"],
             df_fuel_engine_types["Gas"])
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.stackplot(df_fuel_engine_types["Year"],
             df_fuel_engine_types["Diesel"],
             df_fuel_engine_types["Petrol"],
             df_fuel_engine_types["Gas"])
plt.show()
```

```python
colors=["#011638","#7e2987", "#ef2026"]
labels=["Diesel","Petrol","Gas"]
plt.figure(figsize=(12,6))
plt.stackplot(df_fuel_engine_types["Year"],
             df_fuel_engine_types["Diesel"],
             df_fuel_engine_types["Petrol"],
             df_fuel_engine_types["Gas"],
             colors=colors,
             edgecolor='none')
plt.xticks(df_fuel_engine_types["Year"],rotation=45)
plt.legend(labels=labels,loc="upper left")
plt.ylabel("Number of cars",fontsize=13)
plt.title("Popularity of Engine Fuel Types (1982-2016)", fontsize=14, weight="bold")
sns.despine()
plt.show()
```

# Line Chart

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_spx_ftse_00_10 = pd.read_csv("line_chart_data.csv")
df_spx_ftse_00_10
```

```python
df_spx_ftse_00_10["new_date"] = pd.to_datetime(df_spx_ftse_00_10["Date"])
df_spx_ftse_00_10["new_date"]
```

```python
labels = ["S&P 500", "FTSE 100"]
plt.figure(figsize = (20, 8))
plt.plot(df_spx_ftse_00_10["new_date"].values, df_spx_ftse_00_10["GSPC500"].values)
plt.plot(df_spx_ftse_00_10["new_date"].values, df_spx_ftse_00_10["FTSE100"].values)
plt.title("S&P vs FTSE Returns (2000 - 2010)", fontsize = 14, fontweight = "bold")
plt.ylabel("Returns")
plt.xlabel("Date")
plt.legend(labels = labels, fontsize = "large")
plt.show()
```

```python
df_spx_ftse_H2_08 = df_spx_ftse_00_10[(df_spx_ftse_00_10.new_date >= '2008-07-01') &
                                      (df_spx_ftse_00_10.new_date <= '2008-12-31')]
df_spx_ftse_H2_08
```

```python
plt.figure(figsize = (20, 8))
plt.plot(df_spx_ftse_H2_08["new_date"].values, df_spx_ftse_H2_08["GSPC500"].values, color = "midnightblue")
plt.plot(df_spx_ftse_H2_08["new_date"].values, df_spx_ftse_H2_08["FTSE100"].values, color = "crimson")
plt.title("S&P vs FTSE Returns (H2 2008)", fontsize = 14, fontweight = "bold")
plt.ylabel("Returns")
plt.xlabel("Date")
plt.legend(labels = labels, fontsize = "large")
plt.show()
```

# Histogram

# Scatter Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_real_estate = pd.read_csv("scatter_data.csv")
df_real_estate
```

```python
plt.figure(figsize=(12,8))
plt.scatter(df_real_estate['Area (ft.)'],
           df_real_estate['Price'])
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.scatter(df_real_estate['Area (ft.)'],
           df_real_estate['Price'],
           alpha=0.6,
           c=df_real_estate["Building Type"])
plt.show()
```

```python
plt.figure(figsize=(12,8))
scatter=plt.scatter(df_real_estate['Area (ft.)'],
                       df_real_estate['Price'],
                       alpha=0.6,
                       c=df_real_estate["Building Type"],
                       cmap='viridis')
plt.legend(*scatter.legend_elements(),
          loc="upper left",
          title="Building Type")
plt.title("Relationship between Area and Price of California Real Estate",
         fontsize=14,
         weight="bold")
plt.xlabel("Area (sq. ft)", weight="bold")
plt.ylabel("Price (000's of $)")
plt.show()
```

### Scatter plot using seaborn

```python
plt.figure(figsize = (12, 8))
sns.scatterplot(x=df_real_estate['Area (ft.)'],
                y=df_real_estate['Price'],
                hue = df_real_estate['Building Type'],
                palette = ['black', 'darkblue', 'purple', 'pink', 'white'],
                s = 100)
plt.title("Relationship between Area and Price of California Real Estate", 
          fontsize = 14, 
          weight = "bold")
plt.xlabel("Area (sq. ft.)", weight = "bold")
plt.ylabel("Price (000's of $)")
plt.show()
```

# Regression Scatter Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_ad_expenditure=pd.read_csv("scatter_plot_ii.csv")
df_ad_expenditure
```

```python
sns.regplot(x="Budget",
           y="Sales",
           data=df_ad_expenditure)
plt.show()
```

```python
# plt.figure(figsize=(10,8))
sns.set(rc={'figure.figsize':(9,6)})
sns.regplot(x="Budget",
           y="Sales",
           data=df_ad_expenditure)
plt.show()
```

```python
# plt.figure(figsize=(10,8))
sns.set(rc={'figure.figsize':(9,6)})
sns.regplot(x="Budget",
           y="Sales",
           data=df_ad_expenditure,
           color="grey")
plt.show()
```

```python
# plt.figure(figsize=(10,8))
sns.set(rc={'figure.figsize':(9,6)})
sns.regplot(x="Budget",
           y="Sales",
           data=df_ad_expenditure,
           scatter_kws={'color':'k'},
           line_kws={'color':'red'})
plt.xlabel("Ad Expenditure in (000's $)")
plt.ylabel("Sales in (000's Units)")
plt.title("Effect of Ad Expenditure on Sales", fontsize = 14, weight = "bold")
plt.show()
```

```python
sns.lmplot(x = "Budget", 
           y = "Sales", 
           data = df_ad_expenditure)
plt.show()
```

```python
sns.lmplot(x = "Budget", 
           y = "Sales", 
           data = df_ad_expenditure, 
           height = 10,
           scatter_kws = {'color': 'k'},
           line_kws = {'color': 'red'}) 
plt.xlabel("Ad Expenditure in (000's $)")
plt.ylabel("Sales in (000's Units)")
plt.title("Effect of Ad Expenditure on Sales", fontsize = 14, weight = "bold")
plt.show()
```

# Bar and line chart

```python
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib.ticker import PercentFormatter
```

```python
df_kdnuggets = pd.read_csv("bar_line_chart_data.csv")
df_kdnuggets
```

```python
fig, ax=plt.subplots(figsize=(10,7))
ax.bar(df_kdnuggets["Year"],
      df_kdnuggets["Participants"],
      color="k")
```

```python
fig, ax=plt.subplots(figsize=(10,7))
ax.bar(df_kdnuggets["Year"],
      df_kdnuggets["Participants"],
      color="k")
ax1 = ax.twinx()
ax1.plot(df_kdnuggets["Year"].values,
         df_kdnuggets["Python Users"].values, 
         color = "#b60000", 
         marker = "D")
```

```python
fig, ax = plt.subplots(figsize = (10, 7))
ax.bar(df_kdnuggets["Year"],
       df_kdnuggets["Participants"],
       color = "k")
ax.set_ylabel("Number of Participants", 
              weight = "bold")
ax.tick_params(axis = "y", 
               width = 2, 
               labelsize = "large")
ax1 = ax.twinx()
ax1.set_ylim(0, 1)
ax1.yaxis.set_major_formatter(PercentFormatter(xmax = 1.0))
ax1.plot(df_kdnuggets["Year"].values,
         df_kdnuggets["Python Users"].values, 
         color = "#b60000", 
         marker = "D")
ax1.set_ylabel("Python Users", 
               color = "#b60000", 
               weight = "bold")
ax1.tick_params(axis = "y", 
                colors = "#b60000", 
                width = 2, 
                labelsize = "large")
ax.set_title("KD Nuggets Survey Python Users (2012 - 2019)", fontsize = "14", weight = "bold")
plt.show()
```
